```python
column = df["worldwide gross"]
# Sample the data for efficiency
column = column.sample(n=100, random_state=1, replace=True)
input_text = " | ".join(column.astype(str).apply(lambda x: x.split("[")[0]))
input_text = "Transformers | The Dark Knight | The Hobbit: An Unexpected Journey | Avatar | Spider-Man | Transformers: Dark of the Moon | The Lion King | The Incredibles | Titanic | Batman Begins | The Hunger Games | Harry Potter and the Deathly Hallows Part 2 | Furious 7 | The Matrix Reloaded | Superman Returns | Skyfall | Jurassic World | Pirates of the Caribbean: At World's End | Deadpool | The Twilight Saga: Eclipse | Iron Man 3 | Frozen | Pirates of the Caribbean: Dead Man's Chest | The Lord of the Rings: The Return of the King | The Avengers"
import warnings
# Suppress all warnings
warnings.filterwarnings("ignore")

import pandas as pd
import spacy
nlp = spacy.load('en_core_web_trf')
doc = nlp(input_text)
entity_counts = {'LOCATION': 0, 'ORGANIZATION': 0, 'PERSON': 0}
# Define which labels to exclude from 'OTHER' to prevent misclassification
excluded_labels = {'DATE', 'TIME', 'MONEY', 'PERCENT', 'QUANTITY', 'CARDINAL', 'ORDINAL'}

for ent in doc.ents:
    label = ent.label_
    if label == 'PERSON':
        entity_counts['PERSON'] += 1
    elif label == 'ORG':
        entity_counts['ORGANIZATION'] += 1
    elif label in ['GPE', 'LOC']:
        entity_counts['LOCATION'] += 1
    else:
        if label not in entity_counts:
            entity_counts[label] = 1
        entity_counts[label] += 1    
    # Entities in excluded_labels are treated as literals and not counted 
    #print(f'{ent.text} ({ent.label_})')
entity_counts
```

```python
def is_number(value: str) -> bool:
    """
    Check if a value is a number by attempting to convert it to float.

    Parameters:
    - value (str): The value to check.

    Returns:
    - bool: True if it's a number, False otherwise.
    """
    try:
        float(value)
        return True
    except ValueError:
        return False
    

sum(column.apply(is_number))
```
